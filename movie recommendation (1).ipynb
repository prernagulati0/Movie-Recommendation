```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.imdb.com/search/title/?title_type=feature,tv_movie&release_date=1900-01-01,2019-12-31&count=250'
response = get(url)
print(response.text[:300])
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

```python
movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
print(type(movie_containers))
print(len(movie_containers))
```

```python
# Lists to store the scraped data in

names = []
years = []
imdb_ratings = []
metascores = []
Genre = []
```

```python
# Extract data from individual movie container
for container in movie_containers:
# If the movie has Metascore, then extract:
    if container.find('div', class_ = 'ratings-metascore') is not None:
# The name
        name = container.h3.a.text
        names.append(name)
        
# The Genre        
        genre = container.find('span','genre')
        Genre.append(genre)
# The year
        year = container.h3.find('span', class_ = 'lister-item-year').text
        years.append(year)
# The IMDB rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
        
# The Metascore
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))

```

```python
test_df = pd.DataFrame({'movie': names, 'genre' : Genre , 'year': years,'imdb': imdb_ratings,'metascore': metascores})
test_df.sort_values("imdb", axis=0, ascending=False, inplace=True,  na_position='first')
pd.set_option('display.max_rows',None)
test_df
```

```python
l=float(input("Enter the rating above the movie you wanna watch: "))
w=input("enter the year of the movie you wanna watch: ")
a=input("enter the genre of the mmovie you wanna watch: ")
print("recommended movies:")
df2 = test_df.loc[((test_df['imdb'] > l) | (test_df['year']==w)  | (test_df['genre']==a))]
df2
```
