```python
#import modules
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Import packages and set urls
url = 'https://www.imdb.com/search/title/?title_type=feature,tv_movie&release_date=1900-01-01,2020-12-31&count=250'
res = get(url)
print(res.text[0:300])
```

```python
html_soup = BeautifulSoup(res.text, 'html.parser')
type(html_soup)
```

```python
movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
print(type(movie_containers))
print(len(movie_containers))
```

```python
# Lists to store the scraped data in
names = []
years = []
imdb_ratings = []
metascores = []
genres=[]
```

```python
# Extract data from individual movie container
for container in movie_containers:
# If the movie has Metascore, then extract:
    if container.find('div', class_ = 'ratings-metascore') is not None:
# The name
        name = container.h3.a.text
        names.append(name)
        
# The Genre        
        genre = container.find('span','genre')
        genres.append(genre.text.rstrip().replace("\n","").split(","))
# The year
        year = container.h3.find('span', class_ = 'lister-item-year').text
        years.append(year)
# The IMDB rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
        
# The Metascore
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
```

```python
test_df = pd.DataFrame({'movie': names,'genre': genres, 'year': years,'imdb': imdb_ratings,'metascore': metascores})
test_df.sort_values("imdb", axis=0, ascending=False, inplace=True,  na_position='last')
test_df['year']=[x.replace('(',"") for x in test_df['year']]
test_df['year']=[x.replace(')',"") for x in test_df['year']]
test_df['year']=[x.replace('I',"") for x in test_df['year']]
test_df['year']=[x.replace('II',"") for x in test_df['year']]
test_df['year']=[x.replace('III',"") for x in test_df['year']]
test_df['year']=[int(x) for x in test_df['year']]
pd.set_option('display.max_rows', None)
test_df
```
